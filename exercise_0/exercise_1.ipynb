# 1. Product extractor

In this task you will based on product descriptions, be able to extract out relevant information into structured format using pydantic model together with PydanticAI agent.


## a) Read the products.json and store all of its descriptions into a list
## a) Läs in products.json och spara alla descriptions i en lista

```python
import json

with open("data/products.json", "r", encoding="utf-8") as f:
    products = json.load(f)

descriptions = [product["description"] for product in products]
```

## b) Loop through this list and extract structured output from each of the descriptions. The structure should have the fields: name, price, category, in_stock and description

- `name`
- `price`
- `category`
- `in_stock`
- `description`

## b) b) Loopa igenom listan och extrahera strukturerad output från varje beskrivning


```python
from pydantic import BaseModel, Field

class Product(BaseModel):
    name: str = Field(description="Name of the product")
    price: float = Field(description="Price of the product")
    category: str = Field(description="Category of the product")
    in_stock: bool = Field(description="Whether the product is in stock")
    description: str = Field(description="Short description of the product")
```

```python
from pydantic_ai import Agent
from constants import MODEL_LARGE

product_extractor_agent = Agent(
    MODEL_LARGE,
    system_prompt="""
    You are an expert at extracting product information from product descriptions.
    Based on a product description, extract the relevant information.
    Make sure the output matches the schema exactly.
    """
)
```

```python
result = await product_extractor_agent.run(
    descriptions[0],
    output_type=Product
)

result.output
```

```python
result.output.category
```

```python
result.output.price
```

```python
result.output.in_stock
```

```python
result.output.description
```

```python
products_extracted = []

for desc in descriptions:
    result = await product_extractor_agent.run(
        desc,
        output_type=Product
    )
    products_extracted.append(result.output)
```

## c) Now try to validate programmatically if the output fields are correct.

## c) Försök nu att validera programmatiskt om output-fälten är korrekta

```python
validation_results = []

for product in products_extracted:
    result = {
        "name_is_str": isinstance(product.name, str) and len(product.name) > 0,
        "price_is_number": isinstance(product.price, (int, float)) and product.price >= 0,
        "category_is_str": isinstance(product.category, str) and len(product.category) > 0,
        "in_stock_is_bool": isinstance(product.in_stock, bool),
        "description_is_str": isinstance(product.description, str) and len(product.description) > 0
    }
    validation_results.append(result)
```
